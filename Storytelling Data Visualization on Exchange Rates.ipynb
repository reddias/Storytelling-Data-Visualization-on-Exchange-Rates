# Storytelling Data Visualization on Exchange Rates

The dataset we are about to analyze describes daily euro exchange rates between 1999 and 2022. In most countries of the European Union, the Euro (symbolized by €) is the official currency.

Daria Chemkaeva compiled the dataset and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020). The data source is the European Central Bank. Please note that the data set is updated on a regular basis - it was downloaded on January 2023.

```python
import pandas as pd 
exchange_rates = pd.read_csv('D:/euro-daily-hist_1999_2022.csv')
```

## Data Cleaning

Before we start creating data visualizations, we'll need to do a bit of data cleaning. Our focus in the part of the project will be on the exchange rate between the euro and the American dollar.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop = True, inplace=True)
exchange_rates.head()
```

To keep everything clean and easy to work with, let's separate only the columns we need and assign them to the variable `euro_to_dollar`.

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

As you can see above, there are 62 dashes in the `US_dollar` column that should not be there. Once the dashes are removed, it is a good idea to convert this column to a float datatype.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'], linewidth = 3)
plt.show()
```

Instead of seeing a smooth line, when we look at the shape of the line, we see a lot of little wiggles. These wiggles are important: they visually represent how the exchange rate varies from day to day. The rate goes up and down, up and down again, from day to day. Only in the longer term (months or years) does the rate show a clear upward or downward trend.

We may not want to show this daily variation on our graph, depending on our goals. To mask it and show only the long-term trend, we can use a rolling average, called a moving average.

Since the rolling window can have different values, we will use the most common value - 30 (one month).

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'], linewidth = 3)
plt.show()
```

## Coming up with an idea

When thinking about storytelling data visualization, a few ideas came to mind, but the most interesting seem to be the following:

Show comparatively how the euro-dollar rate changed under three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), Donald Trump (2017-2021)). We can use a line plot.

## Storytelling Data Visualization

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & 
                                         (euro_to_dollar['Time'].dt.year <= 2021)]

bush = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year < 2009)]
obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
import matplotlib.style as style
```

```python
style.use('fivethirtyeight')

plt.figure(figsize = (12, 6))
ax1 = plt.subplot(2,3,1) 
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)

axes = [ax1, ax2, ax3, ax4]

### Changes to all the subplots

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)

### Ax1: Bush

ax1.text(x = 12270, y = 1.9, weight = 'bold', size = 18,
         s = 'BUSH', color = 'purple', alpha = 0.5)
ax1.text(x = 12000, y = 1.8, weight = 'bold',
         s = '(2001-2009)', color = 'grey', alpha = 0.5)

ax1.plot(bush['Time'], bush['rolling_mean'],
         color = 'purple', alpha = 0.5)

ax1.set_xticklabels(['','2001','', '2003','','2005','', '2007','', '2009'],
                   alpha=0.3)

### Ax2: Obama

ax2.text(x = 15170, y = 1.9, weight = 'bold', size = 18,
         s = 'OBAMA', color = 'orange', alpha = 0.7)
ax2.text(x = 15040, y = 1.8, weight = 'bold',
         s = '(2009-2017)', color = 'grey', alpha = 0.5)

ax2.plot(obama['Time'], obama['rolling_mean'],
         color = 'orange', alpha = 0.7)

ax2.set_xticklabels(['', '2009','','2011', '','2013','','2015', '','2017'],
                   alpha=0.3)

### Ax3: Trump

ax3.text(x = 17670, y = 1.9, weight = 'bold', size = 18,
         s = 'TRUMP', color = '#00B2EE')
ax3.text(x = 17585, y = 1.8, weight = 'bold',
         s = '(2017-2021)', color = 'grey', alpha = 0.5)

ax3.plot(trump['Time'], trump['rolling_mean'],
         color = '#00B2EE', alpha = 1)

ax3.set_xticklabels(['2017','','2018','', '2019','','2020','','2021'],
                   alpha=0.3)

### Ax4: Bush-Obama-Trump

ax4.plot(bush['Time'], bush['rolling_mean'],
         color = 'purple', alpha = 0.5)
ax4.plot(obama['Time'], obama['rolling_mean'],
         color = 'orange', alpha = 0.7)
ax4.plot(trump['Time'], trump['rolling_mean'],
         color = '#00B2EE', alpha = 1)

ax4.set_xticks([])

### Adding title, subtitle and signature bar

ax.text(10700, 3.4, weight = 'bold', size = 23,
        s = 'EURO-USD rate averaged 1.22 under three US presidents')

ax.text(10700, 3.19, size = 16,
        s = '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),\nand Donald Trump (2017-2021)''')

ax.text(10700, 0.65,
        '©Made by Saidjon Khaydar-Zade' + ' '*102 +
        'Source: European Central Bank',
        color = '#f0f0f0',
        backgroundcolor = '#4d4d4d',
        size=12)

plt.savefig('D:/euro_to_dollar.svg')
plt.show()
```
